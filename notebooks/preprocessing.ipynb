```python
import pandas as pd
import numpy as np
```

```python
log = pd.read_csv("../data/logstore_standard_log_anonymized_no_duplicates.csv")
print(log.isna().sum())
log.head()
```

```python
# Drop unused columns 
log = log.drop(columns=['other'])

# Extract id from username
get_id = lambda x : int(x[4:])
log['username'] = log['username'].apply(get_id)

mask1 = ~log['relatedusername'].isna()
mask2 = ~log['realusername'].isna()
log.loc[mask1, 'relatedusername'] = log.loc[mask1, 'relatedusername'].apply(get_id)
log.loc[mask2, 'realusername'] = log.loc[mask2, 'realusername'].apply(get_id)
log['relatedusername'] = log['relatedusername'].astype('Int64')
log['realusername'] = log['realusername'].astype('Int64')
log = log.rename(columns={
    "username": "userid",
    "relatedusername": "relateduserid",
    "realusername": "realuserid"
})

# Change objectid to INT
log['objectid'] = log['objectid'].astype("Int64")
log['objectid'].isna()
```

```python
log.to_csv("../data/standard_log.csv", index=False)
```
